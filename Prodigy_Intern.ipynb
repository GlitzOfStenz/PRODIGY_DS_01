```python
import pandas as pd
df=pd.read_csv('census2011.csv')
```

```python
df.head()
```

```python
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
continuous_variable=['Population','Growth','Sex-Ratio','Literacy']
categorical_variable=['District','State']
#barchart for categorical variable
for col in categorical_variable:
    if len(df[col].value_counts()) > 20:
        df[col].value_counts().nlargest(15).plot(kind='bar', edgecolor='black', color='green')
    else:
        df[col].value_counts().plot(kind='bar', edgecolor='black', color='green')
    plt.title(f'{col.capitalize()} Distribution')
    plt.xlabel(col.capitalize())
    plt.ylabel('Count')
    plt.xticks(rotation=90)
    plt.show()

```

```python
import matplotlib.pyplot as plt
continuous_variable=['Population','Growth','Sex-Ratio','Literacy']
for col in continuous_variable:
    plt.hist(df[col], bins=10, color='green', edgecolor='black')
    plt.title(f'{col.capitalize()} Distribution')
    plt.xlabel(col.capitalize())
    plt.ylabel('Frequency')
    plt.show()
```
